```python
df = df.drop(columns=['PatientId', 'AppointmentID'])
```

```python
df = df[(df["Age"] > 0)].reset_index(drop=True)
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].map({'M' : 0, 'F' : 1})
df["No-show"] = df["No-show"].map({"No": 0, "Yes": 1})
```

```python
df.head()
```

```python
df['Handcap'].value_counts()
```

```python
df["Days2Appt"] = (pd.to_datetime(df["AppointmentDay"]).dt.day_of_year - pd.to_datetime(df["ScheduledDay"]).dt.day_of_year)
df = df.drop(columns=['ScheduledDay', 'AppointmentDay'])
```

```python
df['Scholarship'].value_counts()
```

```python
df = df[df['Days2Appt'] >= 0].reset_index(drop=True)
```
